### Cleaning Data

```python
import pandas as pd
```

```python
# Try different encodings
try:
    df = pd.read_csv('DataCoSupplyChainDataset.csv', encoding='utf-8')
except UnicodeDecodeError:
    df = pd.read_csv('DataCoSupplyChainDataset.csv', encoding='latin-1')  # or other encodings
```

```python
df.head()
```

```python
#rename column 
df = df.rename(columns={"Customer City": "Department city","Customer Country": "Department country","Customer State": "Department state"})
```

```python
import pandas as pd


# Replace column names if they are different in your actual DataFrame
columns = ['Department city', 'Department state', 'Department country', 'Department Id']

# Create a new DataFrame with only the relevant columns
df_regions = df[columns]

# Drop duplicate rows to get unique combinations of department details
unique_regions = df_regions.drop_duplicates()

# Count the number of unique regions
num_regions = unique_regions.shape[0]

print(f"The number of regions is: {num_regions}")
```

```python
# print orderID count before dropping 
order_id_counts = df['Order Id'].value_counts()
print(order_id_counts)
```

```python
#remove duplicated in order ID
df = df.drop_duplicates(subset='Order Id', keep='first')
order_id_counts_new = df['Order Id'].value_counts()
print(order_id_counts_new)
```

```python
unique_regions
```

```python
# Check for null values and remove
# import pandas as pd

# Check for null values
print(df.isnull().sum())

# Remove rows with null values in any column
# df = df.dropna()
```

```python
# import pandas as pd

# Drop rows with null values in a specific column 
df = df.dropna(subset=['Customer Zipcode'])
```

```python
# Check for null values
print(df.isnull().sum())
```

```python
# import pandas as pd

# Assuming df is your DataFrame
df['Order Item Quantity'] = df['Order Item Quantity'] * 100
```

```python
df.info()
```

```python
# import pandas as pd

# Assuming df is your DataFrame and 'Order Item Quantity' is the name of your column
first_100_values = df['Order Item Quantity'].head(100)
print(first_100_values)
```

```python
# Group by Department Id and get the unique combinations of region details for each department
department_regions = df_regions.groupby('Department Id').agg({
    'Department city': pd.Series.nunique,
    'Department state': pd.Series.nunique,
    'Department country': pd.Series.nunique
})

# Rename the columns for clarity
department_regions.columns = ['Num Cities', 'Num States', 'Num Countries']

# Reset the index to make 'Department Id' a regular column
department_regions.reset_index(inplace=True)

# Display the result
print(department_regions)
```

```python
# Check for duplicates in the specified column
duplicates = df[df.duplicated(subset='Order Id')]

# Drop duplicates from the specified column
df = df.drop_duplicates(subset='Order Id', keep='first')

# Display the updated DataFrame
df.info()
```
