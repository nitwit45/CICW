```python
import random
import pandas as pd
from deap import base, creator, tools, algorithms

# Your list of places
places = [
    ("Place 1", (40.60063934, -73.76037598)),
    ("Place 2", (37.95482636, -122.332962)),
    ("Place 3", (34.02620316, -84.34632111)),
]

# Sample DataFrame with order information
order_data = pd.DataFrame({
    'OrderID': [1, 2, 3],
    'OrderLatitude': [35.6895, 37.95482636, 40.60063934],
    'OrderLongitude': [139.6917, -122.332962, -73.76037598],
    'ShippingMode': ['standard', 'second class', 'first class'],
    'CustomerSegment': ['Consumer', 'Corporate', 'Home Office'],
    'OrderItemQuantity': [10, 5, 8],
    'PreparationTime': [2, 1, 3],
    'OrderDiscount': [0.1, 0.05, 0.2],
    'PaymentMethod': ['Debit', 'Bank Transfer', 'Debit'],
    'ProductCategory': ['Electronics', 'Books', 'Electronics']
})


# Constants for weightages
discount_above_200_weightage = 0.1
discount_below_200_penalty = 0.1
Debit_payment_weightage = 0.4
bank_transfer_payment_weightage = 0.6
home_office_weightage = 0.3
corporate_weightage = 0.4
consumer_weightage = 0.3

# Lead time optimization
inventory_threshold = 100
order_volume_increase_percentage = 0.1
debit_payment_lead_time_weightage = 0.6
bank_transfer_payment_lead_time_weightage = 0.4



# Function to shuffle the order of places in an individual
def shuffle_order():
    places_indices = list(range(len(places)))
    random.shuffle(places_indices)
    PlaceIndex = places_indices[0]
    shipping_mode = random.choice(['standard', 'second class', 'first class'])
    shipping_days = random.randint(1, 10)
    
    return [PlaceIndex,shipping_mode, shipping_days]

# Define DEAP types and fitness function
creator.create("FitnessMulti", base.Fitness, weights=(-1.0,1.0))  # Minimize both distance and cost
creator.create("Individual", list, fitness=creator.FitnessMulti)

# Create a DEAP toolbox
toolbox = base.Toolbox()
toolbox.register("individual", tools.initIterate, creator.Individual, shuffle_order)
toolbox.register("population", tools.initRepeat, list, toolbox.individual)
toolbox.register("mate", tools.cxTwoPoint)
toolbox.register("mutate", tools.mutShuffleIndexes, indpb=0.2)
toolbox.register("select", tools.selTournament, tournsize=3)

def calculate_distance(place1, place2):
    lat1, lon1 = place1
    lat2, lon2 = place2
    return sqrt((lat2 - lat1)**2 + (lon2 - lon1)**2)

# Function to evaluate an individual for a specific order
def evaluate(individual):

    # total_distance, total_cost = 0, 0
    
    # # Iterate over each order
    # for i, order in enumerate(order_info_list):
    #     # Extract relevant information from the individual array
    #     place_indices = individual[:-2]
    #     shipping_mode = individual[-2]
    #     shipping_days = individual[-1]
        
    #     # Retrieve coordinates for the current order
    #     order_coordinates = [(places[j][1][0], places[j][1][1]) for j in place_indices]
        
    #     # Calculate total distance for the current order
    #     for k in range(len(order_coordinates) - 1):
    #         total_distance += calculate_distance(order_coordinates[k], order_coordinates[k + 1])
        
    #     # Calculate cost for the current order based on quantity, discount, and shipping mode weight
    #     order_cost = order['OrderItemQuantity'] * (1 - order['OrderDiscount']) * shipping_mode_weights[shipping_mode]
    #     total_cost += order_cost
        
    # return total_distance, total_cost

    total_distance, total_cost, total_lead_time, total_profit = 0, 0, 0, 0
    
    # Iterate over each order
    for i, order in enumerate(order_info_list):
        # Extract relevant information from the individual array
        place_indices = individual[:-2]
        shipping_mode = individual[-2]
        shipping_days = individual[-1]
        
        # Retrieve coordinates for the current order
        order_coordinates = [(places[j][1][0], places[j][1][1]) for j in place_indices]
        
        # Calculate total distance for the current order
        for k in range(len(order_coordinates) - 1):
            total_distance += calculate_distance(order_coordinates[k], order_coordinates[k + 1])

        def calculate_shipping_mode_cost(shipping_mode):
            cost_map = {'standard': 1, 'second class': 0.8, 'first class': 0.5}
            return cost_map.get(shipping_mode, 1)

    def calculate_shipping_cost(product_category, total_distance):
        weight_category = next(key for key, value in categories_by_weight.items() if product_category in value)
        weight_oz = shipping_rates_data['Weight (oz)'][weight_category]

        zone = determine_usps_zone(total_distance, usps_zones)
        shipping_rate = shipping_rates_data[zone][weight_category]
        shipping_weigtage_mode = calculate_shipping_mode_cost(shipping_mode)
        shipping_cost = weight_oz * shipping_rate * shipping_weigtage_mode
        return shipping_cost

    def determine_usps_zone(total_distance, usps_zones):
        for zone, (lower_bound, upper_bound) in usps_zones.items():
            if upper_bound is None or lower_bound <= total_distance <= upper_bound:
                return zone


        # Calculate shipping cost based on product category and distance
        # product_category = order['ProductCategory']
        # weightage = shipping_cost_by_category.get(product_category, 1)
        # shipping_cost = weightage * total_distance
        shipping_cost=calculate_shipping_cost( order['ProductCategory'],total_distance)
        total_cost += shipping_cost

        # Lead time optimization
        preparation_time = order['PreparationTime']
        order_volume = order['OrderItemQuantity']

        # Consider order volume for lead time optimization
        if order_volume > inventory_threshold:
            preparation_time *= (1 + order_volume_increase_percentage)

        # Consider payment method for lead time optimization
        if order['PaymentMethod'] == 'Debit':
            total_lead_time += preparation_time * debit_payment_lead_time_weightage
        else:
            total_lead_time += preparation_time * bank_transfer_payment_lead_time_weightage

        # Profit optimization
        product_price = order['ProductPrice']
        discount = order['OrderDiscount']
        payment_method = order['PaymentMethod']
        customer_segment = order['CustomerSegment']

        # Discount weightage based on quantity
        if order_volume > 200 and discount > 0:
            discount_weightage = discount_above_200_weightage
        else:
            discount_weightage = -discount_below_200_penalty

        # Payment method weightage
        if payment_method == 'Debit':
            payment_method_weightage = Debit_payment_weightage
        else:
            payment_method_weightage = bank_transfer_payment_weightage

        # Customer segment weightage
        if customer_segment == 'Home Office':
            customer_segment_weightage = home_office_weightage
        elif customer_segment == 'Corporate':
            customer_segment_weightage = corporate_weightage
        else:
            customer_segment_weightage = consumer_weightage

        # Calculate profit per order
        profit_per_order = (product_price - (shipping_cost + discount_weightage)) * (1 + payment_method_weightage)

        # Apply customer segment weightage
        total_profit += profit_per_order * customer_segment_weightage

    return total_distance, total_profit

# toolbox.register("evaluate", evaluate)
# Define the order information as a list of dictionaries
order_info_list = order_data.to_dict(orient='records')
# Register the evaluation function in the toolbox
# toolbox.register("evaluate", evaluate,order_info_list)
toolbox.register("evaluate", evaluate)
# Number of orders
num_orders = len(order_info_list)

# Number of individuals in the population
population_size = 50

# Number of generations
generations = 100

# Loop through each order, create a population, and perform genetic algorithm
# for order_index in range(num_orders):
# f
gen_best_fitness_values = []
best_fitness_values = []
# Loop through each order, create a population, and perform genetic algorithm
for order_index in range(num_orders):
    
    # Create a population for the current order
    population = toolbox.population(n=population_size)

    # Evaluate the entire population for the current order
    fitness_values = list(map(toolbox.evaluate, population))

    # # Check lengths of fitness values and weights
    # print("Length of fitness_values:", len(fitness_values[0]))  # Assuming fitness_values is a list of tuples
    # print("Length of weights in individual:", len(population[0].fitness.weights))

    # Assign fitness values to individuals
    for ind, fit in zip(population, fitness_values):
        ind.fitness.values = fit

    # Assign fitness values to individuals
    # for ind, fit in zip(population, fitness_values):
    #     ind.fitness.values = fit

    # Perform genetic algorithm for the current order
    algorithms.eaMuPlusLambda(population, toolbox, mu=population_size, lambda_=2*population_size,
                              cxpb=0.7, mutpb=0.0, ngen=generations, stats=None, halloffame=None,
                              verbose=False)

    # Find the best individual in the final population
    best_individual = tools.selBest(population, k=1)[0]

    print(best_individual)

    # [0, 'standard', 6]

    # best_individual = 

    # # # Print the results for the current order
    # print(f"Order {order_index + 1} - Best Coordinates: {coordinates}, Best Place Indices: {place_indices}, "
    #     f"Best Shipping Mode: {shipping_mode}, Fitness Values: {best_individual.fitness.values}")
  
    gen_best_fitness_values.append(best_individual.fitness.values)

    # Store the best fitness values for the current order
    best_fitness_values.append(gen_best_fitness_values)

# Plot the convergence plot
# Plot the convergence plot
# Find the minimum length among all fitness value lists
min_length = min(len(fit) for fit in best_fitness_values)

# Plot the convergence plot
for order_index, gen_best_fitness_values in enumerate(best_fitness_values):
    # Trim the lists to the minimum length
    trimmed_fitness_values = gen_best_fitness_values[:min_length]
    plt.plot(range(1, min_length + 1), trimmed_fitness_values, label=f"Order {order_index + 1}")

plt.xlabel('Generation')
plt.ylabel('Best Fitness Value')
plt.title('Convergence Plot')
plt.legend()
plt.show()



# # Extract the coordinates, place indices, and shipping mode from the best individual
# coordinates = [(places[int(place_idx)][1][0], places[int(place_idx)][1][1]) for place_idx in best_individual[:-2]]
# place_indices = best_individual[:-2]
# shipping_mode = best_individual[-2]

# # Print the results for the current order
# print(f"Order {order_index + 1} - Best Coordinates: {coordinates}, Best Place Indices: {place_indices}, "
#       f"Best Shipping Mode: {shipping_mode}, Fitness Values: {best_individual.fitness.values}")

```
